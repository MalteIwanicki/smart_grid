```python
from datetime import datetime,timedelta
import pandas as pd
FILE="../data/lastprofil_h0i_2022_viertelstÃ¼ndig.csv"
```

```python
df = pd.read_csv(FILE).set_index("Datum")
```

```python
from time import strftime, time


(datetime.strptime("01.01.2022 00:15 Sa"[:-3],"%d.%m.%Y %H:%M")-timedelta(minutes=15)).strftime("%Y-%m-%d %H:00:00")
```

```python
usage={}
for date, row in df.iterrows():
    date = (datetime.strptime(date[:-3],"%d.%m.%Y %H:%M")-timedelta(minutes=15)).strftime("%Y-%m-%d %H:00:00")
    usage.setdefault(date,0)
    usage[date]+=row["Verbrauch"]*1000
usage = {date:usage/1000 for date, usage in usage.items()}
usage =[{"date":date,"usage":verbrauch}
for date, verbrauch in usage.items()]
    
usage
```

```python
df_hourly = pd.DataFrame.from_dict(usage)
df_hourly
```

```python
df_hourly.to_parquet("../data/verbrauchsdata_2022_hourly.parquet")
```
