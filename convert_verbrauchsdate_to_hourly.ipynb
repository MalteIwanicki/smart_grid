```python
import pandas as pd
FILE="data/verbauchsdata_1999.xlsx"
```

```python
df = pd.read_excel(FILE).set_index("TagesNr.")
```

```python
df
```

```python
consum_daily={}
for day, row in df.iterrows():
    day_consum={}
    for hour, values in enumerate(zip(row[2::4], row[3::4], row[4::4], row[5::4])):
        day_consum[f"{hour}"]=sum(values)
    consum_daily[day]=day_consum
df_hourly = pd.DataFrame.from_dict(consum_daily,orient="index").rename_axis("day").rename_axis("hour",axis="columns")[:365]
df_hourly.to_parquet("data/verbrauchsdata_1999_hourly.parquet")
```
