```python
import pandas as pd
FILE="data/verbauchsdata_1999.xlsx"
```

```python
df = pd.read_excel(FILE).set_index("TagesNr.")
df
```

```python
df = df.drop(366) #empty
```

```python
from datetime import datetime,timedelta
usage={"date":[], "usage":[]}
for day, row in df.iterrows():
    date = row["Datum"]
    for hour, values in enumerate(zip(row[2::4], row[3::4], row[4::4], row[5::4])):
        usage["date"].append(date+timedelta(hours=hour))
        usage["usage"].append(sum(values))
df_hourly = pd.DataFrame.from_dict(usage)
df_hourly
```

```python
df_hourly.to_parquet("data/verbrauchsdata_1999_hourly.parquet")
```
